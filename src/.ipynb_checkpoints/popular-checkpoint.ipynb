```python
from modules import parse_songs as parser
from modules import util

def parse_artist(artist):
    # 文法構造を解析
    res = parser.parse(f"./json/2_edited/popular/{artist}.json")

    # 文法構造の可視化
    for song_name in res["songs"]:
        parser.visualize(res["songs"][song_name], song_name)
    
    # 曲ごとの構文木のJSONファイルを出力
    for song_name in res["songs"]:
        data = parser.make_tree(res["songs"][song_name])

        dir_path = f"./json/3_output/popular/{artist}"
        util.make_dir(dir_path)

        util.output_json(f"{dir_path}/{song_name}.json", data)
```

```python
parse_artist('GReeeeN')
```
